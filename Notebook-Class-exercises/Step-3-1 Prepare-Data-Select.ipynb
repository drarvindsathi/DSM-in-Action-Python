### 3-1-1 Read files
- Read google_mobility_data_2020
- Read google_mobility_data_2021
- Read google_mobility_data_2022
- Concat the three files to create google_mobility_data

```python
df_google_mobility_data_2020 = pd.read_csv(os.path.join(dir_input, 
                                                   "Google", 
                                                   "Region_Mobility_Report_CSVs", 
                                                   "2020_US_Region_Mobility_Report.csv"))
df_google_mobility_data_2020
```

```python
df_google_mobility_data_2021 = pd.read_csv(os.path.join(dir_input, 
                                                   "Google", 
                                                   "Region_Mobility_Report_CSVs", 
                                                   "2021_US_Region_Mobility_Report.csv"))
df_google_mobility_data_2021
```

```python
df_google_mobility_data_2022 = pd.read_csv(os.path.join(dir_input, 
                                                   "Google", 
                                                   "Region_Mobility_Report_CSVs", 
                                                   "2022_US_Region_Mobility_Report.csv"))
df_google_mobility_data_2022
```

```python
df_google_mobility_data = pd.concat([df_google_mobility_data_2020, 
                                     df_google_mobility_data_2021, 
                                     df_google_mobility_data_2022], ignore_index=True)
df_google_mobility_data
```

```python
df_google_mobility_data.columns
```

### 3-1-2 Select
- sub_region_1
- sub_region_2
- census_fips_code
- date
- Mobility indicators:
-- retail_and_recreation_percent_change_from_baseline
-- grocery_and_pharmacy_percent_change_from_baseline
-- parks_percent_change_from_baseline
-- transit_stations_percent_change_from_baseline
-- workplaces_percent_change_from_baseline
-- residential_percent_change_from_baseline


```python
df_google_mobility_data_selected = df_google_mobility_data[['sub_region_1','sub_region_2','census_fips_code','date',
                                                             'retail_and_recreation_percent_change_from_baseline',
                                                             'grocery_and_pharmacy_percent_change_from_baseline',
                                                             'parks_percent_change_from_baseline',
                                                             'transit_stations_percent_change_from_baseline',
                                                             'workplaces_percent_change_from_baseline',
                                                             'residential_percent_change_from_baseline']]
                                              
df_google_mobility_data_selected
```

```python
df_google_mobility_data_selected = df_google_mobility_data_selected.rename(columns={"census_fips_code": "countyFIPS"})
df_google_mobility_data_selected
```

```python
df_google_mobility_data_selected.columns
```

### 3-1-3 Change data types

Change FIPS code to string after dropping an accidental .0 (see 56045.0 in the last line) added by Python reader
By converting data type to int, it drops the .0 and then it is converted to string

```python
df_google_mobility_data_selected = df_google_mobility_data_selected.dropna(subset=['countyFIPS'])
df_google_mobility_data_selected = df_google_mobility_data_selected.astype({'countyFIPS': int})
df_google_mobility_data_selected = df_google_mobility_data_selected.astype({'countyFIPS': str})
df_google_mobility_data_selected
```

```python
df_google_mobility_data_selected.to_csv(os.path.join(dir_output,  
                                                   "Google_Mobility_Report.csv"))
```
